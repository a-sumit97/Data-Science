#Abstraction

- Abstract class will have atleast one abstract method
- inherit from abstract class
- will have the security method in child class
- you can't create an object of abstract class

```python
from abc import ABC, abstractmethod # abstract base class
class Phone(ABC):

  def database(self):
    print("connected to database")

  @abstractmethod
  def security(self):
    pass

class SmartPhone(Phone):

  def display(self):
    print("inside display")

  def security(self):
    print("inside security")

s = SmartPhone()
s.database()
s.display()
s.security()
```
