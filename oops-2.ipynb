```
class TwoD:

    def __init__(self,x,y):
        self.xcor = x
        self.ycor = y

    def __str__(self):
        return '<{},{}>'.format(self.xcor,self.ycor)

    def euclidean_method(self,other):
        return ((self.xcor-other.xcor)**2 + (self.ycor-other.ycor)**2)**0.5

    def distance_from_origin(self):
        return self.euclidean_method(TwoD(0,0))

class line:

    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c

    def __str__(self):
        return '{}x + {}y + {} = 0'.format(self.a,self.b,self.c)

    def point_on_line(line,TwoD):
        if line.a * TwoD.xcor + line.b * TwoD.ycor + line.c == 0:
            return "lies on the line"
        else:
            return 'dont lies on line'

    def shortest_distance(line,TwoD):
        return abs(line.a * TwoD.xcor + line.b * TwoD.ycor + line.c)/ (line.a**2 + line.b**2)**0.5

    def check_intersect(line,other):
        if line.a * other.b != line.b * other.a:
            return 'lines intersect'
        else :
            return 'dont intersect'

obj1 = TwoD(1,1)
print(obj1)
obj2 = TwoD(-2,-1)
print(obj2)

# print(obj1.euclidean_method(obj2))
# print(obj1.distance_from_origin())



l1 = line(3,4,7)
print(l1.point_on_line(obj1))

l2 = line(1,1,-2)
print(l2.shortest_distance(obj1))

print(l1.check_intersect(l2))
```

#Reference Variable

- Reference variable holds the object
- we can create objects without reference varibale as well
- an object can have multiple reference varibale
- assigning a new reference variable to an existing object does not create a new object

```
#Object without a reference

class person:

  def __init(self):
    self.name = 'sumit'
    self.country = 'India'
```

```
p = person()
q = p
```

```
# Multiple Ref

print(id(p))
print(id(q))
```

# Pass By Reference

```
class Person:

  def __init__(self,name,country):
    self.name = name
    self.country = country

def greet(person): # outside the class -> Function
  print('hello', person.name,'you are from', person.country)
  p1 = Person('madri','USA')
  return p1

p = Person('sumit','India')
x = greet(p)
print(x.name, x.country)
```
