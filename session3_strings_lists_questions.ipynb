```python
# Find the length of a given string without using the len() function

strr = input("enter your string")

counter = 0

for i in strr:
  counter = counter + 1

print(counter)
```

```python
# Extract username from a given email.
# Eg if the email is sumit@gmail.com
# then the username should be sumit

strr = input("enter your email")

ind = strr.index('@')

print(strr[0:ind])
```

```python
# Count the frequency of a particular character in a provided string.
# Eg 'hello how are you' is the string, the frequency of h in this string is 2.

strr = input("enter your string")

to_find = input("enter what you need to find")

counter = 0

for i in strr:
  if i == to_find:
    counter = counter + 1

print(counter)
```

```python
# Write a program which can remove a particular character from a string.
s = input('enter the string')
term = input('what would like to remove')

result = ''

for i in s:
  if i != term:
    result = result + i

print(result)
```

```python
# Write a program that can check whether a given string is palindrome or not.
# abba
# malayalam

s = input("enter your string")

flag = True
for i in range(0,len(s)//2):
  if s[i] != s[len(s) - i -1]:
    flag = False
    print("not a palindrome")
    break
if flag:
  print("palindrome")
```

```python
# Write a program to count the number of words in a string without split()

s = input("enter your string")

lst = []

temp = ''

for i in s:
  if i != ' ':
    temp = temp + i
  else:
    lst.append(temp)
    temp = ''
lst.append(temp)
print(lst)
```

```python
# Write a python program to convert a string to title case without using the title()

s = input('enter your string')

lst = []

for i in s.split():
  lst.append(i[0].upper()  + i[1:].lower())

print(' '.join(lst))
```

```python
# Write a program that can convert an integer to string.

number = int(input('enter your number'))

digit ='0123456789'

strr = ''

while number != 0:
  strr = digit[number%10] + strr
  number = number // 10

print(strr)
print(type(strr))
```
