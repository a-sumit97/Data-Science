### Some Theory

##### Types of data used for I/O:
- Text - '12345' as a sequence of unicode chars
- Binary - 12345 as a sequence of bytes of its binary equivalent

##### Hence there are 2 file types to deal with
- Text files - All program files are text files
- Binary Files - Images,music,video,exe files

### How File I/O is done in most programming languages

- Open a file
- Read/Write data
- Close the file

##Writing to a file

```python
#case 1 if file is not present

f = open('test.txt','w')
f.write("Hello My Name is Sumit")
f.close()
#since file is closed it will not work
f.write("sumit")
```

```python
#write multiline strings

f = open('sample.txt','w')
f.write("Hello My Name Is\n")
f.write("Sumit Rawat\n")
f.write("I'm good")
f.close()
```

```python
# case 2 - if file is already present

f = open('/content/sample.txt','w')
f.write('salman khan')
f.close()
```

```python
#problem with w mode
# introducing append mode

f = open('/content/sample.txt','a')
f.write('is a good hero')
f.close()
```

```python
# write Lines

L = ['hello\n','how are you\n','sumit rawat\n',"i'm good"]

f = open('/content/sample.txt','w')
f.writelines(L)
f.close()
```

```python
#reading from files
# using -> read()

f = open('/content/sample.txt','r')
print(f.read())
f.close()
```

```python
#reading upto n char

f = open('/content/sample.txt','r')
print(f.read(10))
f.close()
```

```python
#readline - to read line by line

f = open('/content/sample.txt','r')
print(f.readline(),end = '')
print(f.readline(),end = '')
print(f.readline(),end = '')

f.close()
```

```python
#reading entire using readline

f = open('/content/sample.txt','r')

while True :

  data = f.readline()

  if data == '':
    break
  print(data,end = '')

f.close()
```

### Using Context Manager (With)

- It's a good idea to close a file after usage as it will free up the resources
- If we dont close it, garbage collector would close it
- with keyword closes the file as soon as the usage is over

```python
#with

with open('/content/sample.txt','w') as f:
  f.write('hello')
```
