```python
# menu driven atm

Balance = 10000
Account_Number = 123456
Pin = 1234

# Login attempt loop (3 chances)
for attempt in range(3, 0, -1):
    acc = int(input("Enter your account number: "))
    pin = int(input("Enter your PIN: "))
    if acc == Account_Number and pin == Pin:
        print("Welcome")
        while True:  # Main menu loop
            menu = input("""\nHello, How can I help you?
1. Check Balance
2. Change Pin
3. Withdraw Cash
4. Deposit Cash
5. Exit\n""")
            if menu == '1':
                print("Your Balance is", Balance)
            elif menu == '2':
                new_pin = int(input("Enter your new PIN: "))
                Pin = new_pin
                print("Pin changed successfully, your new PIN is", Pin)
            elif menu == '3':
                withdraw = int(input("Enter the amount to withdraw: "))
                if withdraw <= Balance:
                    Balance -= withdraw
                    print("Please collect your cash", withdraw)
                    print("Your Balance is", Balance)
                else:
                    print("Insufficient Balance")
            elif menu == '4':
                deposit = int(input("Enter the amount to deposit: "))
                Balance += deposit
                print("Your amount is deposited successfully. New balance is", Balance)
            elif menu == '5':
                print("Thank you! Visit again.")
                break  # Exit menu loop
            else:
                print("Invalid option. Try again.")
        break  # Exit login loop as successful login already happened
    else:
        print("Incorrect Account Number or PIN.")
        if attempt == 1:
            print("Try again tomorrow. Login attempts exceeded.")
        else:
            print(f"{attempt-1} attempts left.")
```

```python
#guessing game

import random

jackpot = random.randint(1,100)

gender = input("enter your gender")

if gender == 'male':

  chances = 5

else :

  chances = 10

i = 1

while i < chances+1:

  guess = int(input("enter your guess"))
  if guess == jackpot:
    print("you won")
    break
  elif guess < jackpot:
    print("guess higher")
  else:
    print("guess lower")
  i += 1
  chances -= 1

  if chances == 0:
    print("you lost")

```
