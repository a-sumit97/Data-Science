#Tuple

Same as list but is immutable.

- Ordered
- Immuatable
- Allows Duplicates

```python
#Create tuple

tup = ()
print(tup)
print(type(tup))

#create a tuple with single item

tup = (1,) # put comma at last to create a single element tuple
print(tup)
print(type(tup))

# homogenous
tup = (1,2,3,4,5)
print(tup)

#heterogenous

tup = (1,'hello',True)
print(tup)

tup = tuple([1,2,3,4])
print(tup)

tup = tuple("hello")
print(tup)
```

#Difference btw Lists and Tuples

- Syntax
- Mutability
- Speed
- Memory
- Built in function
- Error prone
- Usuability

```python
# Tuple take less processing time than list
import time

L = list(range(100000000))
T = tuple(range(100000000))

start = time.time()
for i in L:
    i*5
print("List time: ",time.time() - start)

start = time.time()
for i in T:
  i*5
print("Tuple time: ",time.time() - start)
```

```python
# Tuple take less size in memory

import sys

L = list(range(1000))
T = tuple(range(1000))

print("List size: ",sys.getsizeof(L))
print("Tuple size: ",sys.getsizeof(T))

```

```python
# List are more error prone than tuple

a = [1,2,3,4]
b = a

a.append(5)

print(a)
print(b)
```

```python
# Tuple

a = (1,2,3,4)
b = a

a = a + (5,)

print(a)
print(b)
```

```python
# Tuple Unpacking

a,b,c = (1,2,3)

print(a)
print(b)
print(c)
```

```python
a,b = (1,2,3)

print(a)
print(b)
```

```python
# swap values

a = 1
b = 2

a,b = b,a

print(a,b)
```

```python
a,b,*others = (1,2,3,4)

print(a,b)
print(others)
```

```python
# Zipping tuples

a = (1,2,3,4)
b = (5,6,7,8)

tuple(zip(a,b))
```

#Set

Unordered Collection of elements. Unique and immutable(values can't change) but we can add and delete items.

operations like union, intersection, symmetric difference

Characterstics:

- Unordered
- Mutable
- No duplicates
- Can't contain mutable data types

```python
# Creating Sets

sets = {} # it creates dictionary

print(sets)
print(type(sets))

s = set() # creating empty set

print(s)
print(type(s))

# 1D Sets

sets = {1,2,3,4}

print(sets)

# sets_2d = {1,2,{3,4}} # you can't have it

# print(sets_2d)

# Homo and Hetrogenous

s3 = {1,'hello',True,1.2} # true is considered as 1 that's why true is not printed

print(s3)

s4 = {1,'hello',False,1.2,(1,2,3)}

print(s4)

# s5 = {1,'hello',False,1.2,[1,2,3]} # not allowed

# print(s5)
```

```python
# using type conversion

s = set([1,2,3,4,1])

print(s)
```

```python
# Duplicates not allowed

s = set([1,2,3,4,1,2,3])

print(s)
```

```python
# set can't have mutable items

s = set([1,2,3,4,[1,2,3]])

print(s)
```

```python
s1 = {1,2,3}
s2 = {3,1,2}

print(s1 == s2)
```

#Set Operations

```python
s1 = {1,2,3,4}
s2 = {3,4,5,6}

#union

s3 = s1.union(s2)

# s1 | s2

print(s3)

# intersection

s4 = s1.intersection(s2)

# s1 & s2

print(s4)

# difference

s5 = s1.difference(s2)

# s1 - s2

print(s5)

# symmetric difference

s6 = s1.symmetric_difference(s2)

# s1 ^ s2

print(s6)

# membership test

s7 = {1,2,3,4}

print(1 in s7)
```

# Set Functions

```python
#len/min/max/sorted/sum

s = {1,2,3,4}

len(s)

min(s)

max(s)

sorted(s)

sum(s)
```

```python
# union/update

s1 = {1,2,3,4}
s2 = {5,6,7,8}

s1 | s2

s1.update(s2)

print(s1)
```

```python
# intersection/ intersection_update

s3 = {1,2,3,4}
s4 = {3,4,5,6,7,8}

s3.intersection(s4)

s3.intersection_update(s4)
print(s3)
```

```python
#difference/difference_update

s5 = {1,2,3,4}
s6 = {3,4,5,6}

s5.difference(s6)

s5.difference_update(s6)

print(s5)
```

```python
# symmetricdifference/symmetric_diff_update

s7 = {1,2,3,4}
s8 = {3,4,5,6}

s7.symmetric_difference(s8)

s7.symmetric_difference_update(s8)

print(s7)
```

```python
#isdisjoint/issubset/issuperset

s9 = {1,2,3,4}
s10 = {3,4,5,6}

s9.isdisjoint(s10)

s9.issubset(s10)

s9.issuperset(s10)
```

```python
# copy

s11 = {1,2,3,4}

s12 = s11.copy()

print(s12)
```
