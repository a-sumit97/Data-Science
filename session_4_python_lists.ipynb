## Creating a List

```python
# Empty
print([])
# 1D -> Homo
print([1,2,3,4,5])
# 2D
print([1,2,3,[4,5]])
# 3D
print([[[1,2],[3,4]],[[5,6],[7,8]]])
# Hetrogenous
print([1,True,5.6,5+6j,'Hello'])
# Using Type conversion
print(list('hello'))
```

## Accessing Items from a List

```python
# Indexing
L = [[[1,2],[3,4]],[[5,6],[7,8]]]
#positive
#print(L[0][0][1])

# Slicing
L = [1,2,3,4,5,6]

print(L[::-1])
```

## Adding Items to a List

```python
# append
L = [1,2,3,4,5]
L.append(True)
print(L)
```

```python
# extend
L = [1,2,3,4,5]
L.extend([6,7,8])
print(L)
```

```python
L = [1,2,3,4,5]
L.append([6,7,8])
print(L)
```

```python
L = [1,2,3,4,5]
L.extend('delhi')
print(L)
```

```python
# insert
L = [1,2,3,4,5]

L.insert(1,100)
print(L)
```

# Editing items in a List

```python
L = [1,2,3,4,5]

# editing with indexing
L[-1] = 500

# editing with slicing
L[1:4] = [200,300,400]

print(L)
```

## Deleting items from a List

```python
# del
L = [1,2,3,4,5]

# indexing
del L[-1]

# slicing
del L[1:3]
print(L)
```

```python
# remove

L = [1,2,3,4,5]

L.remove(5)

print(L)
```

```python
# pop
L = [1,2,3,4,5]

L.pop()

print(L)
```

```python
# clear
L = [1,2,3,4,5]

L.clear()

print(L)
```

## Operations on Lists

- Arithmetic
- Membership
- Loop

```python
# Arithmetic (+ ,*)

L1 = [1,2,3,4]
L2 = [5,6,7,8]

# Concatenation/Merge
print(L1 + L2)
```

```python
print(L1*3)
```

```python
L1 = [1,2,3,4,5]
L2 = [1,2,3,4,[5,6]]

print(5 not in L1)
print([5,6] in L2)
```

```python
# Loops
L1 = [1,2,3,4,5]
L2 = [1,2,3,4,[5,6]]
L3 = [[[1,2],[3,4]],[[5,6],[7,8]]]

for i in L3:
  print(i)
```

## List Functions

```python
# len/min/max/sorted
L = [2,1,5,7,0]

print(len(L))
print(min(L))
print(max(L))
print(sorted(L,reverse=True))
```

```python
# count
L = [1,2,1,3,4,1,5]
L.count(5)
```

```python
# index
L = [1,2,1,3,4,1,5]
L.index(1)
```

```python
# reverse
L = [2,1,5,7,0]
# permanently reverses the list
L.reverse()
print(L)
```

```python
# sort (vs sorted)
L = [2,1,5,7,0]
print(L)
print(sorted(L))
print(L)
L.sort()
print(L)
```

```python
# copy -> shallow
L = [2,1,5,7,0]
print(L)
print(id(L))
L1 = L.copy()
print(L1)
print(id(L1))
```
